```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("Electric_Production.csv")
ds
```

```python
ds.columns = ['Date', 'Production']
ds.head()
```

```python
ds.Date=pd.to_datetime(ds.Date)
ds.set_index('Date', inplace = True)
ds.head()
```

```python
ds.plot(figsize = (20,10), linewidth = 5, fontsize = 20)
plt.xlabel('Year', fontsize = 20);
```

```python
ds.rolling(12).mean().plot(figsize =(20,10), linewidth = 5, fontsize = 20)
plt.xlabel('Year', fontsize = 20)
```

```python
ds.diff().plot(figsize = (20,10), linewidth = 5, fontsize = 20)
plt.xlabel('Year', fontsize = 20);
```
